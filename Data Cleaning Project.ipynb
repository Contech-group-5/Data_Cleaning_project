```python
tre_dt_alive=tre_dt[tre_dt['status']=='Alive']
tre_dt_stump_or_dead=tre_dt[(tre_dt['status']=='Dead')|(tre_dt['status']=='stump')]
```

```python
status_alive=tre_dt_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin','25%','75%']]
status_alive
```

```python
tre_dt_alive=tre_dt_alive.merge(status_alive, on='spc_latin',how='left')
tre_dt_alive.rename(columns={'25%':'low'}, inplace=True)
tre_dt_alive.rename(columns={'75%':'ul'}, inplace=True)
tre_dt_alive.head(6)
```

```python
mask=tre_dt_alive['tree_dbh']<tre_dt_alive['low']
tre_dt_alive.loc[mask,'tree_dbh']=tre_dt_alive['low']

mask=tre_dt_alive['tree_dbh']<tre_dt_alive['ul']
tre_dt_alive.loc[mask,'tree_dbh']=tre_dt_alive['ul']
```
